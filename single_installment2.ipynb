### Installment

```python
from absbox import API,EnginePath

localAPI = API(EnginePath.DEV,lang='english',check=False)
```

```python
installment = ["Installment"
                    ,{"originBalance": 1000.0
                    ,"feeRate": ["fix",0.01]
                    ,"originTerm": 12
                    ,"freq": "Monthly"
                    ,"type": "f_p"
                    ,"originDate": "2022-01-01"}
                    ,{"status": "Current"
                        ,"currentBalance":1000
                        ,"remainTerm":8}]
```

```python
r = localAPI.runAsset("2022-01-02"
                     ,[installment]
                     ,poolAssump=("Pool"
                                    ,("Installment", None, None, None, None)
                                    ,None
                                    ,None)
                     ,read=True)

```

#### Installment : Default

```python
r = localAPI.runAsset("2022-01-02"
                     ,[installment]
                     ,poolAssump=("Pool"
                                    ,("Installment", {"CDR":0.01}, None, None, None)
                                    ,None
                                    ,None)
                     ,read=True)

r[0].head()
```

#### Installment: Prepayment

```python
r = localAPI.runAsset("2022-01-02"
                     ,[installment]
                     ,poolAssump=("Pool"
                                    ,("Installment", {"CDR":0.01}, {"CPR":0.01}, None, None)
                                    ,None
                                    ,None)
                     ,read=True)

r[0].head()
```

#### Installment: Fee-Free / Interest Free for first N period

Let's use new `Weekly` frequency 

```python
installment2 = ["Installment"
                    ,{"originBalance": 1200.0
                    ,"feeRate": ["fix",0.01]
                    ,"originTerm": 12
                    ,"freq": "Weekly"
                    ,"type": ("PO_FirstN", 2)
                    ,"originDate": "2022-01-01"}
                    ,{"status": "Current"
                        ,"currentBalance":1200
                        ,"remainTerm":12}]


r = localAPI.runAsset("2021-12-01"
                     ,[installment2]
                     ,read=True)

r[0].head()
```

Same with default with `None`

```python
r = localAPI.runAsset("2021-12-01"
                     ,[installment2]
                     ,poolAssump=("Pool"
                                    ,("Installment", None, None, None, None)
                                    ,None
                                    ,None)
                     ,read=True)

r[0].head()
```
