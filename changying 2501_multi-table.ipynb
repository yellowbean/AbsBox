```python
def multiHeader(x):
    def buildSpan(m, y):
        if m == []:
            m.append((y,1))
            return m
        if y == m[-1][0]:
            m[-1] = (m[-1][0],m[-1][1] + 1)
        else:
            m.append((y,1))
        return m
    return [ reduce(buildSpan, [[]]+list(_)) for _ in transpose(x['columns']) ]
    #return list(transpose(x['columns']))

multiHeader(rs.to_dict(orient="tight"))
```
