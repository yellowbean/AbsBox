```python
from absbox import API,EnginePath
```

```python
localAPI = API(EnginePath.DEV,check=False)
```

```python
myAsset = ["Mortgage"
            ,{"originBalance": 12000.0
            ,"originRate": ["fix",0.045]
            ,"originTerm": 30
            ,"freq": "monthly"
            ,"type": "level"
            ,"originDate": "2021-02-01"}
            ,{"currentBalance": 10000.0
            ,"currentRate": 0.075
            ,"remainTerm": 30
            ,"status": "current"}]

poolPerfAssump = ("Pool",("Mortgage",{"CDR":0.10},None,None,None)
                                    ,None
                                    ,None)
cutoffDate= "2022-03-01"
```

```python
myPool = {'assets':[myAsset],
         'cutoffDate':cutoffDate}

rPool = localAPI.runPool(myPool
                  ,poolAssump=poolPerfAssump
                  ,read=True)
```

```python
rAsset = localAPI.runAsset(cutoffDate
                     ,[myAsset]
                     ,poolAssump=poolPerfAssump
                     ,read=True)
```

```python
rPool['PoolConsol'][0] == rAsset[0]
```
